En este ejercicio exploramos una base de datos sobre niveles de obesidad en personas de Colombia, Perú y México.  

```python
import pandas as pd

df = pd.read_csv("A1.1 Obesidad.csv")

print("Información general del dataset:\n")
print(df.info())

print("\nPrimeras 10 filas:\n")
print(df.head(10))
```

## 2 Nueva variable a considerar
Si yo formara parte del equipo de investigación, me hubiera gustado recopilar el nivel de actividad física (por ejemplo: sedentario, moderado, alto). Esto sería una variable cualitativa ordinal, ya que se clasifica en categorías con un orden lógico.

## 3 Variable de salida
Si yo fuera líder del proyecto y quisiera realizar un estudio de inferencia, definiría como variable de salida el NivelDeObesidad, ya que es la que resume el estado de salud que queremos explicar.  

```python
# Contar cuántas personas pertenecen a cada categoría de NivelDeObesidad
print("Distribución de la variable de salida (NivelDeObesidad):\n")
print(df["NivelDeObesidad"].value_counts())
```

## 4 Gráfica para un solo sexo
En este caso, analizaremos únicamente la población femenina. Mostraremos en el eje Y la variable de salida (NivelDeObesidad) y en el eje X la Edad (variable cuantitativa que nos interesa).  

```python
import matplotlib.pyplot as plt

# Filtrar solo mujeres
df_female = df[df["Sexo"] == "Female"]

plt.figure(figsize=(8,5))
plt.scatter(df_female["Edad"], df_female["NivelDeObesidad"], alpha=0.1, color="purple")
plt.xlabel("Edad")
plt.ylabel("Nivel de Obesidad")
plt.title("Relación entre Edad y Nivel de Obesidad (Mujeres)")
plt.show()
```
